```python
import lightgbm as lgb
import pandas as pd
import numpy as np

def lightgbm_imputation(dataframe, feature):
    df = dataframe.copy()
    
    # Prepare train and test data
    train_data = df[df[feature].notnull()]
    test_data = df[df[feature].isnull()]
    
    if train_data.empty or test_data.empty:
        raise ValueError("Train or test data is empty. Cannot perform imputation.")
    
    X_train = train_data.drop(columns=[feature])
    y_train = train_data[feature]
    X_test = test_data.drop(columns=[feature])

    # Ensure X_test is 2D
    if isinstance(X_test, pd.Series):
        X_test = X_test.to_frame()
    
    # LightGBM parameters
    params = {
        'objective': 'regression',
        'metric': 'rmse',
        'boosting_type': 'gbdt',
        'verbose': -1
    }
    
    train_dataset = lgb.Dataset(X_train, label=y_train)
    model = lgb.train(params, train_dataset, num_boost_round=100)

    # Predict missing values
    predictions = model.predict(X_test)
    df.loc[df[feature].isnull(), feature] = predictions

    return df

# Example usage
df['Age'] = df['Age'].apply(lambda x: np.nan if x == 0 or x > 74 else x)
df = lightgbm_imputation(df, 'Age')
df['Age'] = df['Age'].apply(lambda x: round(x))

```

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import classification_report



X = df.drop(columns=['FraudFound_P']) 
y = df['FraudFound_P']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# تعریف فضای جستجوی ابرپارامترها
param_dist = {
    'n_estimators': [50, 100, 200, 300, 400, 500],        # تعداد درخت‌ها
    'max_depth': [10, 20, 30, None],                     # عمق درخت
    'min_samples_split': [2, 5, 10],                     # حداقل نمونه برای تقسیم گره
    'min_samples_leaf': [1, 2, 4],                       # حداقل نمونه برای یک برگ
    'bootstrap': [True, False],                          # استفاده از bootstrap
}


model = RandomForestClassifier(random_state=42)


random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=50,                # تعداد ترکیب‌های تصادفی که آزمایش می‌شود
    scoring='f1_weighted',    # متریک برای ارزیابی (اینجا F1 weighted)
    cv=5,                     # تعداد fold در Cross-Validation
    verbose=2,                # نمایش جزئیات فرآیند
    random_state=42,
    n_jobs=-1                 # استفاده از همه هسته‌های CPU
)


random_search.fit(X_train, y_train)


print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)


best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
print("\nClassification Report:")
classification_report(y_test, y_pred)
```
